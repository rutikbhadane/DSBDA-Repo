```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("dataset_Facebook.csv",delimiter = ";")
```

```python
df
```

```python
# data subsets
# selecting only first 250 rows 
df1 = df.iloc[0:250,:]
df2 = df.iloc[250:500,:]
```

```python
df1
```

```python
df2
```

```python
# combining data subsets
new_df = pd.concat([df1,df2])
```

```python
new_df
```

```python
# sorting the dataset based on month
df.sort_values(by=["Post Month"],ascending=True)
```
