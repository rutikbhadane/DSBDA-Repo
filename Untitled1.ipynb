```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
import seaborn as sns
```

```python
df = pd.read_csv("AirQuality.csv",sep = ";")
```

```python
df
```

```python
df.head()
```

```python
# calculating null values
df.isna().sum()
```

```python
# dropping columns with 100% null values

df = df.drop(["Unnamed: 15"],axis=1)
df = df.drop(["Unnamed: 16"],axis=1)
```

```python
df
```

```python
df.dtypes
```

```python
df["PT08.S1(CO)"].fillna(df["PT08.S1(CO)"].mode(),inplace=True)
df["NMHC(GT)"].fillna(df["NMHC(GT)"].median(),inplace=True)
df["PT08.S2(NMHC)"].fillna(df["PT08.S2(NMHC)"].median(),inplace=True)
df["NOx(GT)"].fillna(df["NOx(GT)"].mean(),inplace=True)
df["PT08.S3(NOx)"].fillna(df["PT08.S3(NOx)"].median(),inplace=True)
df["NO2(GT)"].fillna(df["NO2(GT)"].mean(),inplace=True)
df["PT08.S4(NO2)"].fillna(df["PT08.S4(NO2)"].median(),inplace=True)
df["PT08.S5(O3)"].fillna(df["PT08.S5(O3)"].median(),inplace=True)
```

```python
df["CO(GT)"].fillna(df["CO(GT)"].ffill(),inplace=True)
```

```python
df["T"] = df["T"].ffill()
```

```python
df["RH"]  = df["RH"].ffill()
```

```python
df["AH"]  = df["AH"].ffill()
```

```python
df["Time"]  = df["Time"].ffill()
```

```python
df["Date"]  = df["Date"].ffill()
```
